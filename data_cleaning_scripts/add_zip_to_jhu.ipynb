```python
import pandas as pd 
import numpy as np 
import matplotlib.dates as mdates
import matplotlib.pyplot as plt
```

```python
jhuconfirmed_us = pd.read_csv('../raw_data/csse_data/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_US.csv')
jhudeaths_us= pd.read_csv('../raw_data/csse_data/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_US.csv')
jhudeaths_us
```

```python
zip_df = pd.read_csv('../raw_data/zipcode_data/zip_code_database.csv')
# match primary_city to county 
counties = zip_df['county']
cities = zip_df['primary_city']
city_to_county = dict(zip(cities, counties)) 
city_to_county
```

```python
jhuconfirmed_us.insert(2, 'county', jhuconfirmed_us['Admin2'].map(city_to_county))
jhudeaths_us.insert(2, 'county', jhudeaths_us['Admin2'].map(city_to_county))
```
