```python
jhu_global = Table.read_table("../raw_data/csse_data/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv")
jhu_global
```

```python
def get_growthRate(city):
    X = np.arange(88)
    X = sm.add_constant(X)
    temp = jhu_global.where("Country/Region", city).row(0)[4:-1]
    Y = np.log(np.add(temp, 1))

    mod = sm.OLS(Y, X)
    res = mod.fit()
    return 10** res.params[1]
```

```python
def get_time_min(city):
    city_t = apple_transit_raw.where("region", city).row(0)[12:]
    city_ts = savgol_filter(city_t, 35, 3)
    return np.argmin(np.gradient(city_ts, 1))
```

```python
temp = apple_transit_raw.column("region")

regions2 = []
for reg in temp:
    if reg in jhu_global.column("Country/Region"):
        regions2.append(reg)
```

```python
growth_rate = []
for region in regions2:
    growth_rate.append(get_growthRate(region))
```
